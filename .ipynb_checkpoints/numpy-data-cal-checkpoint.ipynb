```python
# PCA
import numpy as np 
import matplotlib.pyplot as plt

A1_mean = [1, 1]
A1_cov = [[2, 0.99], [1, 1]]
A1 = np.random.multivariate_normal(A1_mean, A1_cov, 500)   # 多元正态分布


A2_mean = [5, 5]
A2_cov = [[2, 0.99], [1, 1]]
A2 = np.random.multivariate_normal(A2_mean, A2_cov, 500)

B_mean = [5, 0]
B_cov = [[.5, -1], [.9, -.5]]
B = np.random.multivariate_normal(B_mean, B_cov, 500)


f = plt.figure(figsize=(20,20))

ax = f.add_subplot(111)
ax.scatter(A1[:, 0], A1[:, 1], color='r')
ax.scatter(A2[:, 0], A2[:, 1], color='g')
ax.scatter(B[:, 0], B[:, 1], color='b')

from sklearn.decomposition import KernelPCA
kpca = KernelPCA(kernel='cosine', n_components=1)
AB = np.vstack((A1, B))
# print AB.shape
# print AB
AB_transformed = kpca.fit_transform(AB)
# print AB_transformed

# print AB_transformed
A_label = np.array(['r']*len(B))
B_label = np.array(['b']*len(B))
colors = np.hstack((A_label, B_label))
f = plt.figure(figsize=(10, 4))
ax1 = f.add_subplot(211)
ax2 = f.add_subplot(212)
ax1.set_title("Cosine KPCA 1 Dimension")
ax1.scatter(AB_transformed, np.zeros_like(AB_transformed), color=colors)

from sklearn.decomposition import PCA
pca = PCA(1)
AB_transformed_Reg = pca.fit_transform(AB)
ax2.set_title("PCA 1 Dimension")
ax2.scatter(AB_transformed_Reg, np.zeros_like(AB_transformed_Reg), color=colors)
```

```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn import pipeline, preprocessing, decomposition

iris = load_iris()
iris_data = iris.data
np.random.seed(0)
mask = np.random.binomial(1, 0.25, iris_data.shape).astype(bool)
iris_data[mask] = np.nan
print iris_data[:5]


pca = decomposition.PCA()
imputer = preprocessing.Imputer()

pipe = pipeline.Pipeline([('imputer', imputer), ('pca', pca)])
np.set_printoptions(2)   # 设置打印的小数点位数: 两位
iris_data_transformed = pipe.fit_transform(iris_data)
print iris_data_transformed[:5]


print pipe.get_params()
pipe.set_params(pca__n_components=2)
print pipe.get_params()


zjw = pipe.fit_transform(iris_data)
print zjw[:5]
```
